```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from matplotlib import rc
import scienceplots
%config InlineBackend.figure_format = 'retina'
plt.style.use('science')
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
plt.rcParams.update({'font.size':17})
```

```python
def line(x, m, k):
    return m*x+k
```

```python
U = np.array([35, 34, 32, 30, 28, 26, 24, 22])*1e-3
U_inv = 1/U
print(1/(35e-3)**2*1e-4)
l_min = np.array([30.1, 31, 33.5, 36.4, 39.1, 42.1, 46.2, 50.4])*1e-12
popt, cov = curve_fit(line, U_inv, l_min*1e12)
m = popt[0]
m_err = np.sqrt(float(cov[0][0]))
k = popt[1]
k_err = np.sqrt(float(cov[1][1]))
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(U_inv, l_min*1e12, color='black', s=16)
ax.plot(U_inv, m*U_inv+k, color='black')
ax.set_xlabel('Inverse of voltage, V$^{-1}$ (MV$^{-1}$)')
ax.set_ylabel('Minimum wavelength, $\lambda_{min}$ (pm)')
ax.errorbar(U_inv, l_min*1e12, linestyle='none', ecolor='black', capsize=3, yerr=[3, 3.2, 3.4, 3.8, 4.2, 4.6, 4.8, 5])

print(m/(3*1e8)*1.6e-16*1e-9, m_err/(3*1e8)*1.6e-16*1e-9)
plt.show()
```
